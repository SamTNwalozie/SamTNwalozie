## Identifying missing value

```python
df.info()
```

Two common methods to determine values in Panda Dataframe are missing are 'isna'  and 'isnull' . They are exactly the same method but different names.

```python
# Notice we have a Pandas series of True and False values
df['interest_paid'].isna().head()
```

```python
interest_missing = df['interest_paid'].isna()
```

```python
# look at the row that contains the NaN for interest_paid
df.loc[interest_missing,:]
```

```python
# Keep in mind that we can use the not operator (~)to negate the filter.
# every row that doesnt have a NaN is returned.
df.loc[~interest_missing,:]
```

```python
# The code counts the nmuber of missing values
# sum() works cos booleans are a subtype of integers
df['interest_paid'].isna().sum()
```

## Removing or Filling in Missing Data
This is important subject before you graph any data , you should make sure you arent trying to graph some missing values that can cause  errors or misinterpretation of values.

```python
df.info()
```

## Removing missing values
You can remove missing values using "dropna" method

```python
# You can drop the entire row if they contain 'any'nan in them or 'all'
# this may not be the best strategy for our dataset
df[30:40].dropna(how = 'any')
```

## Filling in Missing values
There are variety of ways to fill in missing values.

```python
# Look at where the missing data is located 
df['interest_paid'][30:40]
```

```python
# Filling in the nan with probably a zero is a bad idea.
df['interest_paid'][30:40].fillna(0)
```

```python
# backfill in value
df['interest_paid'][30:40].fillna(method='bfill')
```

```python
# forwardfill in value
df['interest_paid'][30:40].fillna(method ='ffill')
```

```python
# Linear intepolation (filling of values)
df['interest_paid'][30:40].interpolate(method ='linear')
```

```python
# Interest paid before filling in the NaN with a value
df['interest_paid'].sum()
```

```python
# Filling in the actual value
interest_missing = df['interest_paid'].isna()
df.loc[interest_missing,'interest_paid'] = 93.24
```

```python
# Interest paid after filling in the NaN with a value
df['interest_paid'].sum()
```

```python
# Notice we dont have NaN values in the Dataframe anymore.
df.info()
```

## MATLAB style vs Object Syntax

Matplotlib has two types of syntax

## The MATLAB-Style
This is a scripted interface designed to feel like MATLAB. Matplotlib maintains a pointer to the current(active)figure and sends commands to it.

## Object-Oriented 
This is often used in situations where you want more control over your figure.

## Important note: You can and often will have plots that are created through a combination of both

## The MATLAB-Style

```python
plt.plot(month_number, interest_paid, c = 'k')
plt.plot(month_number, principal_paid, c = 'b')
```
