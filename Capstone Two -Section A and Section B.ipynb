# Q1. what are the Players and Fantasy Premier  League(FPL) Ratings accumulated from Previous Season?


```python
 
print(dataframe_epl[["name","club","big_club","position","fpl_points"]].set_index('name').sort_values("fpl_points",ascending=False).head(50))

dataframe_epl[["name","fpl_points"]].set_index('name').sort_values("fpl_points",ascending=False).head(50).plot(kind='bar')

plt.show()
```

It can be seen that the  top Players play for the Big clubs which is assigned a value of one. 

The Visualization shows the Players rated in a descending order according to thier accumulated Fpl points.

# SECTION B: INSIGHTS

Having analysed my dataset , It is my desire to actually find out as a football fanatic if actually there exist a relationship exist between the respective varaiables analysed above. 

# Q1: Is there relationship between Players Popularity and their Market value? 

```python
sns.regplot(x='page_views',  y='market_value',data=dataframe_epl, color="g", truncate=False, scatter=True, fit_reg=True,)
plt.show()
```

The visualization between the market value and page views of the Footballers clearly show that there exist a linear relationship.But there is an outlier observed which is Wayne Rooney who is a well known english player not playing for a big club then there was also lot of interest from big clubs for him like Manchester united because he had just broken a goal scoring record for a young player then , most popular players play for the big clubs in EPL.

# Q2: Is there a relationship between the Market value of players and their Fpl Point?

```python
sns.regplot(x='market_value', y='fpl_points', data=dataframe_epl, marker="+", truncate=False, color="b")
plt.show()
```

There exist a positive relationship between the accumulated Fanstasy Premier league points and the market value.

# Q3: Is there a relationship between the Market value of players and their current Fpl value?

```python
sns.regplot(x='market_value', y='fpl_value', data=dataframe_epl, truncate=False)
plt.show()

```

There exist a positive relationship between the Fanstasy Premier league(fpl) value and the market value of the Footballer.
